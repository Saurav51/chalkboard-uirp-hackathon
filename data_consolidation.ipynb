```python
import pandas as pd
```

```python
gpa = pd.read_csv("datasets/uiuc-gpa-dataset.csv")
gpa.head()
```

```python
gpa.columns
```

```python
math_col = [231, 257]
```

```python
gpa_filtered = gpa[
    ((gpa['Subject'] == 'MATH') & (gpa['Number'].isin(math_col))) |
    ((gpa['Subject'] == 'CS') & (gpa['Number'] == 124)) |
    ((gpa['Subject'] == 'STAT') & (gpa['Number'] == 107)) |
    ((gpa['Subject'] == 'IS') & (gpa['Number'] == 445)) |
    ((gpa['Subject'] == 'ECON') & (gpa['Number'] == 102))
]
```

```python
gpa_filtered
```

```python
# GPA weights dictionary
gpa_weights = {
    'A+': 4.0,
    'A': 4.0,
    'A-': 3.67,
    'B+': 3.33,
    'B': 3.0,
    'B-': 2.67,
    'C+': 2.33,
    'C': 2.0,
    'C-': 1.67,
    'D+': 1.33,
    'D': 1.0,
    'D-': 0.67,
    'F': 0.0
}

for grade in gpa_weights:
    if grade not in gpa_filtered.columns:
        gpa_filtered[grade] = 0

grade_cols = list(gpa_weights.keys())
gpa_filtered['Total_Grade_Points'] = sum(gpa_weights[grade] * gpa_filtered[grade] for grade in grade_cols)
gpa_filtered['Total_Students'] = gpa_filtered[grade_cols].sum(axis=1)

gpa_filtered['Overall_GPA'] = gpa_filtered['Total_Grade_Points'] / gpa_filtered['Total_Students']
```

```python
gpa_filtered
```

```python
gpa_filtered.to_csv("datasets/gpa_filtered.csv")
```

```python
# Define grade buckets
grade_buckets = {
    'A_range': ['A+', 'A', 'A-'],
    'B_range': ['B+', 'B', 'B-'],
    'C_range': ['C+', 'C', 'C-'],
    'D_range': ['D+', 'D', 'D-'],
    'F_range': ['F']
}

# Step 1: Make a copy and compute total students per row if not already done
gpa_filtered = gpa_filtered.copy()
grade_cols = sum(grade_buckets.values(), [])  # flatten the list
gpa_filtered['Total_Students'] = gpa_filtered[grade_cols].sum(axis=1)

# Step 2: Create new percentage columns per row
for bucket, grades in grade_buckets.items():
    gpa_filtered[bucket] = gpa_filtered[grades].sum(axis=1) / gpa_filtered['Total_Students'] * 100


gpa_filtered_prof = gpa_filtered.groupby(['Primary Instructor', 'Course Title', 'Subject', 'Number']).agg({
    'Overall_GPA': 'mean',
    'Total_Students': 'sum',
    'A_range': 'mean',
    'B_range': 'mean',
    'C_range': 'mean',
    'D_range': 'mean',
    'F_range': 'mean',
    'YearTerm': 'count'
}).reset_index()
```

```python
import pandas as pd
import altair as alt

# Load data
df = pd.read_csv('datasets/viz1.csv')

# Create combined column
df['Subject_Number'] = df['Subject'].astype(str) + ' ' + df['Number'].astype(str)

# Create hover tooltip
hover = ['Primary Instructor', 'Overall_GPA', 'A_range', 'B_range', 'C_range',
         'D_range', 'F_range', 'Total_Students', 'YearTerm']

# Dropdown to filter by Subject + Course Number
subject_number_dropdown = alt.binding_select(options=sorted(df['Subject_Number'].unique()))
subject_number_select = alt.selection_single(fields=['Subject_Number'], bind=subject_number_dropdown, name='Course')

# Main chart
chart = alt.Chart(df).mark_bar(size=50).encode(
    x=alt.X('Primary Instructor:N', sort='-y', title='Instructor'),
    y=alt.Y('Overall_GPA:Q', title='Average GPA'),
    color=alt.Color('Overall_GPA:Q', scale=alt.Scale(scheme='blues')),
    tooltip=hover
).add_selection(
    subject_number_select
).transform_filter(
    subject_number_select
).properties(
    width=800,
    height=400,
    title="Instructor GPA Disparity by Course"
)

chart
```
