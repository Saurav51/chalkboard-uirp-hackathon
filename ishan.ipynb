```python
import pandas as pd
import altair as alt

# Load CSV
df = pd.read_csv("/Users/ishandesai/Desktop/UIRP/chalkboard-uirp-hackathon/datasets/viz1.csv")

# Construct course and instructor fields
df['Course'] = df['Subject'] + " " + df['Number'].astype(str)
df['Instructor'] = df['Primary Instructor']

# Dropdown binding
course_list = sorted(df['Course'].unique())
course_dropdown = alt.binding_select(options=course_list, name='Select Course: ')
course_select = alt.selection_point(fields=['Course'], bind=course_dropdown, value=course_list[0])

# Dynamically compute height based on selected course (initial value)
initial_course = course_list[0]
initial_count = df[df['Course'] == initial_course].shape[0]
chart_height = max(1500, initial_count * 45)

# Base chart
base = alt.Chart(df).transform_filter(course_select)

# GPA bar chart
gpa_chart = base.mark_bar().encode(
    x=alt.X('Overall_GPA:Q', title='Average GPA', scale=alt.Scale(domain=[0, 4])),
    y=alt.Y('Instructor:N', sort='-x', title='Instructor'),
    color=alt.Color('Overall_GPA:Q', scale=alt.Scale(scheme='blues')),
    tooltip=[
        alt.Tooltip('Instructor:N'),
        alt.Tooltip('Overall_GPA:Q', title='GPA'),
        alt.Tooltip('Total_Students:Q', title='Students'),
        alt.Tooltip('A_range:Q', title='A %', format=".1f"),
        alt.Tooltip('B_range:Q', title='B %', format=".1f"),
        alt.Tooltip('C_range:Q', title='C %', format=".1f"),
        alt.Tooltip('D_range:Q', title='D %', format=".1f"),
        alt.Tooltip('F_range:Q', title='F %', format=".1f"),
    ]
).add_params(
    course_select
).properties(
    width=700,
    height=chart_height,
    title=' GPA by Instructor (Filtered by Course)'
)

gpa_chart

```

```python
import pandas as pd
import altair as alt

# Load and prepare data
df = pd.read_csv("/Users/ishandesai/Desktop/UIRP/chalkboard-uirp-hackathon/datasets/viz1.csv")
df['Course'] = df['Subject'] + " " + df['Number'].astype(str)
df['Instructor'] = df['Primary Instructor']

# Dropdown binding
course_list = sorted(df['Course'].unique())
course_dropdown = alt.binding_select(options=course_list, name='Select Course: ')
course_select = alt.selection_point(fields=['Course'], bind=course_dropdown, value=course_list[0])

# Filtered chart base
base = alt.Chart(df).transform_filter(course_select)

# --- Dynamically calculate height and width based on instructor count ---
# NOTE: This only works statically at first render unless used in Streamlit

# Choose an initial course (dropdown default)
initial_course = course_list[0]
num_instructors = df[df['Course'] == initial_course].shape[0]

# Dynamic dimensions
chart_height = max(300, num_instructors * 45)
chart_width = 700 if num_instructors > 15 else 500

# GPA bar chart
gpa_chart = base.mark_bar().encode(
    x=alt.X('Overall_GPA:Q', title='Average GPA', scale=alt.Scale(domain=[0, 4])),
    y=alt.Y('Instructor:N', sort='-x', title='Instructor'),
    color=alt.Color('Overall_GPA:Q', scale=alt.Scale(scheme='blues')),
    tooltip=[
        alt.Tooltip('Instructor:N'),
        alt.Tooltip('Overall_GPA:Q', title='GPA'),
        alt.Tooltip('Total_Students:Q', title='Students'),
        alt.Tooltip('A_range:Q', title='A %', format=".1f"),
        alt.Tooltip('B_range:Q', title='B %', format=".1f"),
        alt.Tooltip('C_range:Q', title='C %', format=".1f"),
        alt.Tooltip('D_range:Q', title='D %', format=".1f"),
        alt.Tooltip('F_range:Q', title='F %', format=".1f"),
    ]
).add_params(
    course_select
).properties(
    width=chart_width,
    height=chart_height,
    title='GPA by Instructor (Filtered by Course)'
)

gpa_chart
```
